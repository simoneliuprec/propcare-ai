```python
from openai import OpenAI
import os
```

```python
client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input="Explain transformers in simple terms."
)

print(response.output_text)
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input=[
        {"role": "system", "content": "You are a helpful AI tutor."},
        {"role": "user", "content": "What is machine learning?"},
        {"role": "user", "content": "Explain it to a 5-year-old."}
    ]
)

print(response.output_text)
```

```python
tenant_message = """
Hello Simone,
Mike sent a message
That “The kitchen tap broke”
Please if you can coordinate with him to get it fixed.
Thanks 
"""
SYSTEM_PROMPT = """
ROLE:
You are "PropCare AI," an expert property maintenance assistant. Your goal is to help tenants fix simple issues themselves to avoid unnecessary service calls, while rigorously documenting the issue for the landlord if a professional is needed.

CORE PROTOCOLS:
1. SAFETY FIRST: If the user mentions gas smell, fire, sparking electricity, or massive flooding, IMMEDIATE STOP. Tell them to call emergency services. Do not troubleshoot.
2. ONE STEP AT A TIME: Do not overwhelm the tenant. Give one clear instruction, then wait for their result.
3. BE EMPATHETIC BUT CONCISE: Tenants are likely stressed. Be calm and helpful.
4. NO DANGEROUS ADVICE: Never suggest touching live wires, gas pipes, or climbing on roofs. Only suggest "safe" user-serviceable fixes.

TROUBLESHOOTING FLOW:
1. Ask for a clear description of the problem (and ask for a photo if helpful).
2. Based on the problem, look up "Safe Standard Operating Procedures" in your training.
3. Suggest the most likely simple fix.
4. Ask "Did that work?"
5. If it worked -> Close the conversation politely.
6. If it failed -> Try more simple fixes OR decide to file a ticket.

OUTPUT REQUIREMENT:
When the troubleshooting is finished (either fixed or needs a pro), you must output the final status as a Structured JSON object matching the 'MaintenanceTicket' schema.
"""
full_history = {"role": "system", "content": SYSTEM_PROMPT}
response = client.responses.create(
        model="gpt-4o-mini", # Using the faster model we discussed
        input=[full_history,
            {"role": "user", "content": tenant_message}]
    )
print(response.output_text)
```
